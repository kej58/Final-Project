```python
##import pandas and matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

```python
##import the data from a csv file and display the first 5 rows

facil = pd.read_csv("facil.csv")
facil.head()
```

```python
##use the length function to get the total amount of data points

len(facil)
```

```python
##display the first 5 rows again, but this time have the first collumn display the neighborhood name

facil2 = pd.read_csv("facil.csv",
                    index_col="neighborhood")
facil2.head()
```

```python
##count each type of facility in ascending order
##we can see that shelters are the most dominant at 61

facil['type'].value_counts().sort_values(ascending=True)
```

```python
##create a pie chart to show the least/most facilities another way
##this pie chart is very simple to read, and we can clearly see that shelter is first, dugout is second, and consession is third

plt.rcParams['figure.figsize']=(25,17)
plt.title('Facility', fontweight=20,fontsize=20)
plt.style.use("seaborn")
pie = facil.type.value_counts()
pie.iloc[:30].sort_values().plot.pie()
plt.xticks(rotation=90)
plt.show()
```

```python
##list the number of facilities from each neighborhood
##there are several neighborhoods with only 1 facility, but Highland Park contains the most
##we also get the total number of neighborhoods at the very bottom of the output (75)

hood_list = facil.groupby('neighborhood').size().sort_values(ascending=True)
hood_list
```

```python
##create a query mask to display the first 5 neighborhoods that include shelters

query_mask = facil2['type'] == 'Shelter'

shelter = facil2[query_mask]
shelter.head()
```

```python
##create a query mask to display all of the rec centers, there should be 6

query_mask = facil2['type'] == 'Rec Center'

rec = facil2[query_mask]
rec
```

```python
##since Highland Park contains the most facilities, lets see everything it includes

query_mask = facil['neighborhood'] == 'Highland Park'
Highland = facil[query_mask]
Highland

##pay attention to this collumn:                        (HERE)
```

```python
##count the 10 neighborhoods that include the most facilities

most = facil.groupby('neighborhood').size().sort_values(ascending=False)
most.head(10)
```

```python
##plot these 10 facility-heavy neighborhoods with a bar graph

data = most.head(10).plot(figsize=(20,10),kind='barh',title="NEIGHBORHOODS WITH THE MOST FACILITIES")
plt.show()
```
