## Data

#### **Incident rate**

```python
#import pandas
import pandas as pd
#import matplotlib for graphing
import matplotlib.pyplot as plt
```

```python
#We first import the data from a csv file and display the first 5 rows
blotter_data = pd.read_csv('BlotterData.csv')
blotter_data.head()
```

```python
#count incident that happened in each location and list it in numerical order
blotter_data['INCIDENTLOCATION'].value_counts().sort_values(ascending=True)
```

```python
#Use the matplotlib for graphing
#Graph it as a pie chart to see which incident reported the most
plt.rcParams['figure.figsize']=(20,20)
plt.title('Incident Reported', fontweight=30,fontsize=20)
plt.style.use("seaborn")
#Find out the most reported crime
rate=blotter_data.INCIDENTHIERARCHYDESC.value_counts()
#plot it as a bar graph
rate.iloc[:10].sort_values().plot.pie()
plt.xticks(rotation=90)
plt.show()
```

```python
#report the first 10 neigborhood that report 99
#99 is the emergency service response code
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('99')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#report the first 10 neigborhood that report theft
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('THEFT')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#report the first 10 neigborhood that report simple assault
report = blotter_data[blotter_data.INCIDENTHIERARCHYDESC.str.contains('HIT AND RUN')];
report.INCIDENTNEIGHBORHOOD.value_counts()
report.iloc[:10]
```

```python
#count the first 20 neighborhood with the least crime reported
less_crime=blotter_data.groupby('INCIDENTNEIGHBORHOOD').size().sort_values(ascending=True)
less_crime.head(20)
```

```python
#plot the region with the least incident reported
data = less_crime.head(20).plot(figsize=(20,10),kind='barh',title="Neighborhood with least incident")
plt.show()
```

#### **Facilities**

```python
##import the data from a csv file and display the first 5 rows
facil = pd.read_csv("facil.csv")
facil.head()
```

```python
##use the length function to get the total amount of data points

len(facil)
```

```python
##display the first 5 rows again, but this time have the first collumn display the neighborhood name
facil2 = pd.read_csv("facil.csv",index_col="neighborhood")
facil2.head()
```

```python
##count each type of facility in ascending order
##we can see that shelters are the most dominant at 61

facil['type'].value_counts().sort_values(ascending=True)
```

```python
##create a pie chart to show the least/most facilities another way
##this pie chart is very simple to read, and we can clearly see that shelter is first, dugout is second, and consession is third

plt.rcParams['figure.figsize']=(25,17)
plt.title('Facility', fontweight=20,fontsize=20)
plt.style.use("seaborn")
pie = facil.type.value_counts()
pie.iloc[:30].sort_values().plot.pie()
plt.xticks(rotation=90)
plt.show()
```

```python
##list the number of facilities from each neighborhood
##there are several neighborhoods with only 1 facility, but Highland Park contains the most
##we also get the total number of neighborhoods at the very bottom of the output (75)

hood_list = facil.groupby('neighborhood').size().sort_values(ascending=True)
hood_list
```

```python
##create a query mask to display the first 5 neighborhoods that include shelters

query_mask = facil2['type'] == 'Shelter'

shelter = facil2[query_mask]
shelter.head()
```

```python
##create a query mask to display all of the rec centers, there should be 6

query_mask = facil2['type'] == 'Rec Center'

rec = facil2[query_mask]
rec
```
